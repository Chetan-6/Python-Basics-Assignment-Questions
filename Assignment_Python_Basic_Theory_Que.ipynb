2. What is an interpreter in Python ?
   - An interpreter translates high-level Python code into machine code (understood by the computer) one instruction at a time, and executes it immediately.

3. What are pre-defined keywords in Python ?
   - Pre-defined keywords in Python are reserved words that have special meaning in the language. You cannot use them as variable names, function names, or identifiers.

4. Can keywords be used as variable names ?
   - No, keywords cannot be used as variable names in Python.
   Because Keywords are reserved words in Python that have special meaning and are part of the syntax (like if, while, for, def, class, etc.). Using them as variable names would confuse the interpreter.

5. What is mutability in Python ?
   - Mutability refers to whether or not an object can be changed after it is created.
   
6. Why are lists mutable, but tuples are immutable ?
   - Why lists are mutable in Python:
   > Design Purpose: Lists are designed to be dynamic collections.
   > You can change: You can add, remove, or modify elements without creating a new list.
   -Why tuples are immutable:
   > Design Purpose: Tuples are meant to be fixed collections of items.
   > Safety: Since they can't change, they are safer to use when passing data around — no accidental changes.
   > Hashable: Because they’re immutable, tuples can be used as keys in dictionaries, unlike lists.

   > Use "is" when checking object identity, like x is None.

8. What are logical operators in Python ?
   - Logical operators are used to combine conditional (boolean) statements. They return True or False based on the logic.

9.  What is type casting in Python ?
    - Type casting means converting the data type of a value from one type to another — like turning a string into an integer, or a float into a string.

10. What is the difference between implicit and explicit type casting ?
    > Implicit Type casting -
    Implicit type casting (also called type coercion) is when Python automatically converts one data type into another during an operation — without you writing any conversion code.
    Key Points -
    Happens automatically
    Done safely
    Usually converts to a higher or more general data type (like int → float)
    > Explicit Type casting -
    Explicit type casting is when you manually convert one data type into another using built-in functions like int(), float(), str(), etc.
    This gives you full control over how the data is converted.

11. What is the purpose of conditional statements in Python ?
    - Conditional statements are used to make decisions in your Python program. They let your code choose different paths based on certain conditions (True/False).
    Purpose: To control the flow of execution — so your program can:
    > Do something only if a condition is true.
    > Do something else if it's not.
    > Handle multiple scenarios.

12.  How does the elif statement work ?
     - The elif statement (short for "else if") is used to check multiple conditions after an if statement. It helps your program decide which block of code to run, among several options.

13. What is the difference between for and while loops ?
    - Both for and while loops are used to repeat a block of code, but they are used in different situations.
    >"for" Loop:
    Used when you know in advance how many times you want to loop.
    Typically used for iterating over a sequence (like a list, string, or range).
    > "while" Loop:
    Used when you don't know ahead of time how many times to loop.
    Runs as long as a condition is true.

14. Describe a scenario where a while loop is more suitable than a for     loop ?
    - Example: User Login Authentication
    You want to keep asking the user for a password until they enter the correct one.
    You don’t know how many attempts the user will take — maybe 1, maybe 10.
    This is perfect for a while loop.
    Why "while" is better here:
    You don’t know in advance how many times to loop.
    You want to keep repeating the check until a condition is met.
    Why "for" is not suitable:
    "for" loops are good when you know the number of iterations (e.g., "for i in range(3)").
    In this case, the loop needs to continue indefinitely until a condition changes — which is exactly what "while" is for.
